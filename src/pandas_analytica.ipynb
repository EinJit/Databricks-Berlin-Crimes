```python
import pandas as pd
import numpy as np

df = pd.read_csv('Berlin_crimes.csv')
df.head()
```

```python
df['nonviolent'] = df['Robbery'] + df['Street_robbery'] + df['Theft'] + df['Fire'] + df['Damage'] + df['Graffiti'] + df['Drugs'] + df['Local']
df['violent'] = df['Injury'] + df['Arson'] + df['Agg_assault'] + df['Threat']
df.to_csv('../fixtures/Berlin_crimes_prepared.csv')
df.head()
```

```python
df2 = df.drop(['Robbery', 'Code', 'Street_robbery', 'Car', 'From_car', 'Bike', 'Burglary', 'Injury', 'Agg_assault', 'Threat', 'Theft', 'Fire', 'Damage', 'Graffiti', 'Drugs', 'Local', 'Arson'], axis=1)
df2 = df2.groupby(['Year', 'District']).sum()
df2 = df2.drop(['Location'], axis=1)
df2.head()
```

```python
df2.to_csv('../fixtures/Berlin_crimes_prepared_agg.csv')
df2
```
